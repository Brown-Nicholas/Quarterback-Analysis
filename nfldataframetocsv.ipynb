```python
#Here we import our nfl data api as well as pandas and numpy
import nfl_data_py as nfl
import pandas as pd
import numpy as np
```

```python
#We then pull the weekly and seasonal data from the 2020 nfl season
```

```python
year = 2020
weekly_data = nfl.import_weekly_data([year])
seasonal_data = nfl.import_seasonal_data([year])
```

```python
#Since we are going to focus primarily on the QB position
#We create a weekly data fame from the week by week data, in order to get players names
```

```python
weekly_qb = pd.DataFrame(nfl.import_weekly_data([year]))
sorted_weekly_qb = pd.DataFrame(weekly_qb.sort_values(by='passing_yards'))
```

```python
#Here we pick the stats we will be working with in our analysis
```

```python
qb_stats = ['player_id','completions','attempts','passing_yards','passing_tds','interceptions','sacks','passing_air_yards','passing_yards_after_catch' ]
```

```python
#The next few lines, we pull the specified data from above from the whole season, and clean it up by passing attempts of 150 or more
```

```python
qb_select = seasonal_data[qb_stats]
```

```python
qb_cleaned = qb_select[qb_select['attempts'] > 149]
```

```python
qb_cleaned
```

```python
#The next few lines we order by passing yards, create a data frame and drop duplicates in the weekly data
#We drop the duplicates to avoid errors when creating the player name column in our qb stats dataframe 
```

```python
qb_yards_desc = qb_cleaned.sort_values(by='passing_yards',ascending=False)
```

```python
qb_stats_merg = pd.DataFrame(qb_yards_desc)
```

```python
weekly_qb = weekly_qb.drop_duplicates(subset='player_id',keep='first')
```

```python
weekly_qb
```

```python
#Next, we create a dictionary with player names and id numbers from our weekly data to copy the names to our qb data
#We then merge the names onto the qb data by matching the player ids
#We do a similar merging process for our qbr data, but we just match the last names
```

```python
name_map = dict(zip(weekly_qb['player_id'],weekly_qb['player_name']))
```

```python
qb_stats_merg['player_name'] = qb_stats_merg['player_id'].map(dict(zip(weekly_qb['player_id'], weekly_qb['player_name'])))
```

```python
qb_stats_merg['last_name'] = qb_stats_merg['player_name'].str.split('.').str[-1]
```

```python
qbr_data = pd.DataFrame(nfl.import_qbr([2020]))
```

```python
qb_stats_merg['qbr'] = qb_stats_merg['last_name'].map(dict(zip(qbr_data['name_last'], qbr_data['qbr_total'])))
```

```python
qb_stats_merg = qb_stats_merg.dropna(subset=['player_name'])
```

```python
qb_stats_merg = qb_stats_merg.drop(columns=['last_name'])
```

```python
qb_stats_merg.to_csv(f'qb_data{year}.csv',index=False)
```
